```python
import math

def calculate_distance(coordinates):
  point1, point2 = coordinates
  distance = math.sqrt((point2[0] - point1[0])) ** 2 + (point2[1] - point1[1] ** 2)
  return distance

coordinates_list = [(1, 1), (2, 3)]
distance = calculate_distance(coordinates_list)
print(f"The distance between the points {coordinates_list[0]} and {coordinates_list[1]} is: {distance: .2F}" )
```

```python
phonebook = [

    {'id': 1, 'name': 'John Doe', 'contact': '09060611233', 'address' :'123 Main St'},
    {'id': 2, 'name': 'Jhonatan', 'contact': '09123456789', 'address' :'456 Main St'},
    {'id': 3, 'name': 'Ryan', 'contact': '09874562310', 'address' :'789 Main St'}

]


def display_contact_by_id(phonebook, record_id):
    for record in phonebook:
        if record['id'] == record_id:
           print(f"Nmae: {record['name']}")
           print(f"Contact: {record['contact']}")
           print(f"Address: {record['address']}")
           return
    print("Record not Found")

for i in range (3):
      try:
        user_id = int(input("Enter the ID of the record you want to look up: "))
        display_contact_by_id(phonebook, user_id)

      except ValueError:
        print("Invalid, Input valid ID.")
```
