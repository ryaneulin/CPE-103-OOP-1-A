```python
class Grade:

  def calculate_ave(self, activities, quizzes, assignments):
      hands_on_avg = sum(activities) / len(activities)
      quiz_avg = sum(quizzes) / len(quizzes)
      assignment_avg = sum(assignments) / len(assignments)
      return (hands_on_avg*0.5) + (quiz_avg*0.3) + (assignment_avg*0.2)

  def get_scores(self, activity_count, score_type):
      scores = []
      for i in range(activity_count):
          while True:
              try:
                score = float(input(f"{score_type} {i + 1}: "))
                total_items = float(input("Total Items: "))
                if total_items <= 0:
                    print("Total items cannot be zero. Please enter a valid number.")
                    continue
                if total_items <= 9:
                    print("Total items cannot be less than or equal to 9. Please enter a valid number.")
                    continue
                scores.append((score / total_items) * 100)
                break
              except ValueError:
                   print("Invalid input. Please enter numeric values.")
      return scores


  def ucc_conversion(self, final_grade):
    if final_grade >= 97:
        return 1.00
    elif final_grade >= 94:
        return 1.25
    elif final_grade >= 91:
        return 1.50
    elif final_grade >= 88:
        return 1.75
    elif final_grade >= 85:
        return 2.00
    elif final_grade >= 82:
        return 2.25
    elif final_grade >= 79:
        return 2.50
    elif final_grade >= 76:
        return 2.75
    elif final_grade >= 60:
        return 3.00
    elif final_grade >= 50:
        return 4.00
    elif final_grade >= 40:
        return 5.00
    else:
        return "Failed"


  def calculate_grades(self):

     print("\nPrelim")
     print("")
     activities =self.get_scores(5, "Activity")
     quizzes = self.get_scores(2, "Quiz")
     assignments = self. get_scores(3, "Assignment")
     prelim_CS = self.calculate_ave(activities, quizzes, assignments)
     prelimExam = float(input("Enter Prelim Exam: "))
     self.prelimGrade = (prelimExam*0.5) + (prelim_CS*0.5)
     print("Prelim Grade:", self.prelimGrade)


     print("\nMidterm")
     print("")
     activities =self.get_scores(5, "Activity")
     quizzes = self.get_scores(2, "Quiz")
     assignments = self.get_scores(3, "Assignment")
     midterm_CS = self.calculate_ave(activities, quizzes, assignments)
     midtermExam = float(input("Enter Midterm Exam: "))
     self.midtermGrade = (prelimExam*0.33) + (midterm_CS*0.67)
     print("Midterm Grade:", self.midtermGrade)
     print("")

     print("\nFinal")
     print("")
     activities =self.get_scores(5, "Activity")
     quizzes = self.get_scores(2, "Quiz")
     assignments = self.get_scores(3, "Assignment")
     final_CS = self.calculate_ave(activities, quizzes, assignments)
     finalExam = float(input("Enter Final Exam: "))
     self.finalGrade = (midtermExam*0.33) + (final_CS*0.67)
     print("Final Grade:", self.finalGrade)
     print("")

     converted_grade = self.ucc_conversion(self.finalGrade)
     print(converted_grade, "Converted Grade UCC:")

  def conclusion(self):
     print(f"The students Prelim Grade {self.prelimGrade}, Midterm Grade {self.midtermGrade}, and Final Grade {self.finalGrade}. ")

  print("""


1. Rhovic M. Gabijan
2. Ryan Bertrand B. Eulin
3. Ken R. Manongsong
 "q" to Quit Program

""")




def main():
        while True:
            choice = input("Enter the Number of your choice, q to quit. : ")

            if choice in [ '1', '2', '3' ]:
               student_grade  = Grade()
               student_grade.calculate_grades()
               student_grade.conclusion()

            elif choice in [ "Quit the Program", "Quit", "Q", "q", "quit", "quit the program"]:
                 break
            else:
                  print("Invalid choice. Please enter a valid option.")

if __name__ == "__main__":
    main()
```
