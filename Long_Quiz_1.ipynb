```python

class Student:
  def school_identity1(self):

    student1_id = "20240001"
    student1_name = "Ana"
    student1_program = "BSEE"
    student1_sex = "Female"
    student1_weight = "48kg"
    student1_height = "160cm"

    print("StudentID: ", student1_id)
    print("Student Name: ", student1_name)
    print("Student Program: ",student1_program)
    print("Student Sex: ", student1_sex)
    print("Student Weight: ",student1_weight)
    print("Student Height: ", student1_height)

  def school_identity2(self):

    student2_id = "20240002"
    student2_name = "Mench"
    student2_program = "BSIE"
    student2_sex = "Male"
    student2_weight = "56g"
    student2_height = "172cm"

    print("StudentID: ", student2_id)
    print("Student Name: ", student2_name)
    print("Student Program: ",student2_program)
    print("Student Sex: ", student2_sex)
    print("Student Weight: ",student2_weight)
    print("Student Height: ", student2_height)


  def school_identity3(self):

    student3_id = "20240003"
    student3_name = "John"
    student3_program = "BSCPE"
    student3_sex = "Male"
    student3_weight = "65kg"
    student3_height = "183cm"

    print("StudentID: ", student3_id)
    print("Student Name: ", student3_name)
    print("Student Program: ",student3_program)
    print("Student Sex: ", student3_sex)
    print("Student Weight: ",student3_weight)
    print("Student Height: ", student3_height)

  def student_enrolled(self):

    course1 =  "CPE 103"
    course2 =  "GEC 002"
    course3 =  "NSTP 122"

    print("Course 1: ", course1)
    print("Course 2: ", course2)
    print("Course 3: ", course3)


  def finalgrade_course(self):

    CPE103 = 1.5
    GEC002 = 1
    NSTP122 = 1

    print("-" * 30)
    print("FINAL GRADE PER COURSE\n")
    print("Final Grade for CPE 103: ", CPE103)
    print("Final Grade for GEC 002: ", GEC002)
    print("Final Grade for NSTP 122: ", NSTP122)



  def calculate_ave(self, activities, quizzes, assignments):
      hands_on_avg = sum(activities) / len(activities)
      quiz_avg = sum(quizzes) / len(quizzes)
      assignment_avg = sum(assignments) / len(assignments)
      return (hands_on_avg*0.5) + (quiz_avg*0.3) + (assignment_avg*0.2)

  def get_scores(self, activity_count, total_score):
    scores = []
    for i in range(activity_count):
        while True:
            try:
              score = float(input(f"{total_score} {i + 1}: "))
              total_items = float(input("Total Item: "))
              if total_items <= 0:
                  print("Invalid Character. Input number should be Less Than or Equal to 0, Try Again.")
                  continue
              if total_items < 9:
                  print("Invalid Character. Input number should be Greater Than to 9, Try Again.")
                  continue
              scores.append((score / total_items) * 100)
              break
            except ValueError:
                print("Invalid Input, Try Again.")
    return scores



  def ucc_conversion(self, final_grade):
    if final_grade >= 97:
        return 1.00
    elif final_grade >= 94:
        return 1.25
    elif final_grade >= 91:
        return 1.50
    elif final_grade >= 88:
        return 1.75
    elif final_grade >= 85:
        return 2.00
    elif final_grade >= 82:
        return 2.25
    elif final_grade >= 79:
        return 2.50
    elif final_grade >= 76:
        return 2.75
    elif final_grade >= 60:
        return 3.00
    elif final_grade >= 50:
        return 4.00
    elif final_grade >= 40:
        return 5.00
    else:
        return "Failed"

  def calculate_grades(self):

     print("-" * 30)
     print("\nMidterm")
     print("")
     activities =self.get_scores(5, "Activity")
     quizzes = self.get_scores(2, "Quiz")
     assignments = self.get_scores(3, "Assignment")
     midterm_CS = self.calculate_ave(activities, quizzes, assignments)
     midtermExam = float(input("Enter Midterm Exam: "))
     self.midtermGrade = (midtermExam*0.33) + (midterm_CS*0.67)
     print("Midterm Grade:", self.midtermGrade)
     print("-" * 30)

     print("\nFinal")
     print("")
     activities =self.get_scores(5, "Activity")
     quizzes = self.get_scores(2, "Quiz")
     assignments = self.get_scores(3, "Assignment")
     final_CS = self.calculate_ave(activities, quizzes, assignments)
     finalExam = float(input("Enter Final Exam: "))
     self.finalGrade = (midtermExam*0.33) + (final_CS*0.67)
     print("Final Grade:", self.finalGrade)
     print("")

     converted_grade = self.ucc_conversion(self.finalGrade)
     print(converted_grade, "Converted Grade UCC:")



print("""

Ana
Mench
John
"q" to Quit the program

""")


def main():
     while True:
      choice = input("Enter your choosen STUDENT : ")

      if choice ==  'Ana' :
         student_grade  = Student()
         student_grade.school_identity1()
         student_grade.calculate_grades()
         student_grade.student_enrolled()
         student_grade.finalgrade_course()

      elif choice == 'Mench':
         student_grade  = Student()
         student_grade.school_identity2()
         student_grade.calculate_grades()
         student_grade.student_enrolled()
         student_grade.finalgrade_course()

      elif choice == 'John':
         student_grade  = Student()
         student_grade.school_identity3()
         student_grade.calculate_grades()
         student_grade.student_enrolled()
         student_grade.finalgrade_course()

      elif choice == 'q':
         break

      else:
         print("Invalid choice, Please enter a valid option")
if __name__ == "__main__":
    main()
```
