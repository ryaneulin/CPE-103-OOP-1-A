```python
import random

class Character():
    def __init__(self, username):
        self.__username = username
        self.__hp = 100
        self.__mana = 100
        self.__damage = 5
        self.__str = 0
        self.__vit = 0
        self.__int = 0
        self.__agi = 0

    def getUsername(self):
        return self.__username

    def setUsername(self, new_username):
        self.__username = new_username

    def getHp(self):
        return self.__hp

    def setHp(self, new_hp):
        self.__hp = new_hp

    def getDamage(self):
        return self.__damage

    def setDamage(self, new_damage):
        self.__damage = new_damage

    def getStr(self):
        return self.__str

    def setStr(self, new_str):
        self.__str = new_str

    def getVit(self):
        return self.__vit

    def setVit(self, new_vit):
        self.__vit = new_vit

    def getInt(self):
        return self.__int

    def setInt(self, new_int):
        self.__int = new_int

    def getAgi(self):
        return self.__agi

    def setAgi(self, new_agi):
        self.__agi = new_agi

    def reduceHp(self, damage_amount):
        self.__hp = max(0, self.__hp - damage_amount)

    def addHp(self, hp_amount):
        self.__hp += hp_amount

class Novice(Character):
    def basicAttack(self, character):
        character.reduceHp(self.getDamage())
        print(f"{self.getUsername()} performed Basic Attack! - {self.getDamage()} damage")

class Swordsman(Novice):
    def __init__(self, username):
        super().__init__(username)
        self.setStr(5)
        self.setVit(10)
        self.setHp(self.getHp() + self.getVit())

    def slashAttack(self, character):
        new_damage = self.getDamage() + self.getStr()
        character.reduceHp(new_damage)
        print(f"{self.getUsername()} performed Slash Attack! - {new_damage} damage")

class Archer(Novice):
    def __init__(self, username):
        super().__init__(username)
        self.setAgi(5)
        self.setInt(5)
        self.setVit(5)
        self.setHp(self.getHp() + self.getVit())

    def rangedAttack(self, character):
        new_damage = self.getDamage() + random.randint(0, self.getInt())
        character.reduceHp(new_damage)
        print(f"{self.getUsername()} performed Ranged Attack! - {new_damage} damage")

class Magician(Novice):
    def __init__(self, username):
        super().__init__(username)
        self.setInt(10)
        self.setVit(5)
        self.setHp(self.getHp() + self.getVit())

    def heal(self):
        self.addHp(self.getInt())
        print(f"{self.getUsername()} performed Heal! - {self.getInt()} HP restored")

    def magicAttack(self, character):
        new_damage = self.getDamage() + self.getInt()
        character.reduceHp(new_damage)
        print(f"{self.getUsername()} performed Magic Attack! - {new_damage} damage")

class Monster(Character):
    def __init__(self):
        super().__init__("Monster")
        self.setHp(150)  # Boss has more HP

def select_role(username):
    print("Select your role: 1. Swordsman 2. Archer 3. Magician")
    choice = int(input("Enter choice (1-3): "))
    if choice == 1:
        return Swordsman(username)
    elif choice == 2:
        return Archer(username)
    elif choice == 3:
        return Magician(username)
    else:
        print("Invalid choice, defaulting to Novice.")
        return Novice(username)

def game_loop(player1, player2):
    while player1.getHp() > 0 and player2.getHp() > 0:
        if random.choice([True, False]):
            player1.basicAttack(player2)
        else:
            player2.basicAttack(player1)

        print(f"{player1.getUsername()} HP: {player1.getHp()}")
        print(f"{player2.getUsername()} HP: {player2.getHp()}")

    if player1.getHp() <= 0:
        print(f"{player2.getUsername()} wins!")
    else:
        print(f"{player1.getUsername()} wins!")

def main():
    print("Welcome to the Game!")
    mode = input("Select mode: 1. Single Player 2. Player vs Player: ")

    if mode == '1':
        username = input("Enter your username: ")
        player = Novice(username)
        wins = 0

        while True:
            if wins >= 2:
                player = select_role(username)

            monster = Monster()
            print(f"A wild {monster.getUsername()} appears!")
            game_loop(player, monster)

            if player.getHp() > 0:
                wins += 1
                print(f"You have {wins} wins!")
            else:
                print("Game Over!")
                break

    elif mode == '2':
        username1 = input("Enter username for Player 1: ")
        player1 = select_role(username1)

        username2 = input("Enter username for Player 2: ")
        player2 = select_role(username2)

        game_loop(player1, player2)

if __name__ == "__main__":
    main()
```
